```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import backend as K
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import pickle
```

```python
class Sampling(layers.Layer):
    """Uses (z_mean, z_log_var) to sample z, the vector encoding a digit."""

    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.keras.backend.random_normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon


class Encoder(keras.Model):
    """Maps to a triplet (z_mean, z_log_var, z)."""

    def __init__(self, input_shape,filters,kernels,strides,latent_dim=32, intermediate_dim=64, name="encoder", **kwargs):
        super(Encoder, self).__init__(name=name, **kwargs)
        self._input_shape=input_shape
        self.filters=filters      
        self.kernels=kernels
        self.strides=strides
        self.latent_dim=latent_dim
        self._shape_before_bottleneck=None
        self._num_conv_layers = len(filters)
        self.encoder = self.build_encoder()


    def call(self, inputs):
        x = self.encoder(inputs)

        return x
    
    def _save_parameters(self,folder):
                    
        parameters = [
            self._input_shape,
            self.filters,
            self.kernels,
            self.strides,
            self.latent_dim
        ]
        save_path = os.path.join(folder, "parameters.pkl")
        with open(save_path, "wb") as f:
            pickle.dump(parameters, f)
        
    def build_encoder(self):
        
        #latent_dim = 2
        
        encoder_inputs = keras.Input(shape=self._input_shape, name="encoder_input") 
        x = self._add_conv_layers(encoder_inputs)
        self._shape_before_bottleneck = K.int_shape(x)[1:]
        x = layers.Flatten()(x)
        z_mean = layers.Dense(self.latent_dim, name="z_mean")(x)
        z_log_var = layers.Dense(self.latent_dim, name="z_log_var")(x)
        z = Sampling()([z_mean, z_log_var])
        encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
        #encoder.summary()
        return encoder
    
    def summary(self):
        self.encoder.summary()
        
    def _add_conv_layers(self, encoder_input):
        """Create all convolutional blocks in encoder."""
        x = encoder_input
        for layer_index in range(self._num_conv_layers):
            x = self._add_conv_layer(layer_index,x)
        return x
    
    def _add_conv_layer(self, layer_index,x):
        layer_number = layer_index + 1
        conv_layer = layers.Conv2D(
            filters=self.filters[layer_index],
            kernel_size=kernels[layer_index],
            strides=strides[layer_index],
            padding="same",
             name=f"encoder_conv_layer_{layer_number}"
        )
        x = conv_layer(x)
        x = layers.ReLU(name=f"encoder_relu_{layer_number}")(x)
        return x
    
    def get_shape_before_bottleneck(self):
        return self._shape_before_bottleneck
    
    
class Decoder(keras.Model):
    """Converts z, the encoded vector, back into a readable."""
    # TODO output dim still hard coded
    def __init__(self, filters, latent_dim, shaper,name="decoder", **kwargs):
        super(Decoder, self).__init__(name=name, **kwargs)
        self.filters=filters      
        self.kernels=kernels
        self.strides=strides
        self._num_conv_layers = len(filters)
        self.latent_dim=latent_dim
        self.shaper=shaper
        self.decoder=self.build_temp()


    def call(self, inputs):
        x = self.decoder(inputs)
        return x
    
    def build_temp(self):
        latent_inputs = keras.Input(shape=(self.latent_dim,))
        x = layers.Dense(np.prod(self.shaper), activation="relu")(latent_inputs)
        x = layers.Reshape(self.shaper)(x)
        x = self._add_conv_transpose_layers(x)
        # TODO fix this hardcoding
        decoder_outputs = layers.Conv2DTranspose(1, 3, activation="sigmoid", padding="same")(x)
        decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
        return decoder
    
    def _add_conv_transpose_layers(self, x):
        """Add conv transpose blocks."""
        # loop through all the conv layers in reverse order and stop at the
        # first layer
        for layer_index in reversed(range(self._num_conv_layers)):
            x = self._add_conv_transpose_layer(layer_index, x)
        return x

    def _add_conv_transpose_layer(self, layer_index, x):
        layer_num = self._num_conv_layers - layer_index
        conv_transpose_layer = layers.Conv2DTranspose(
            filters=self.filters[layer_index],
            kernel_size=self.kernels[layer_index],
            strides=self.strides[layer_index],
            padding="same",
            name=f"decoder_conv_transpose_layer_{layer_num}"
        )
        x = conv_transpose_layer(x)
        x = layers.ReLU(name=f"decoder_relu_{layer_num}")(x)
        #x = BatchNormalization(name=f"decoder_bn_{layer_num}")(x)
        return x
    
    def summary(self):
        self.decoder.summary()
```

```python
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.model = self._build_model()
        self.total_loss_tracker = keras.metrics.Mean(name="total_loss")
        self.reconstruction_loss_tracker = keras.metrics.Mean(
            name="reconstruction_loss"
        )
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")
        
    def _build_model(self):
        model_input = layers.Input(shape=self.encoder._input_shape)
        model_output = self.decoder(self.encoder(model_input))
        return keras.Model(model_input, model_output)
        
    def save(self,folder="."):
        # make directory if not exists
        if not os.path.exists(folder):
            os.makedirs(folder)
        
        self.encoder._save_parameters(folder)
        
        save_path = os.path.join(folder, "weights.h5")
        self.model.save_weights(save_path)
    
    @classmethod
    def load(cls, save_folder="."):
        parameters_path = os.path.join(save_folder, "parameters.pkl")
        with open(parameters_path, "rb") as f:
            parameters = pickle.load(f)
        encoder = Encoder(*parameters)
        shaper=encoder.get_shape_before_bottleneck()
        filters=encoder.filters
        latent_dim = encoder.latent_dim
        decoder=Decoder(filters,latent_dim,shaper)
        vae=VAE(encoder,decoder)
        weights_path = os.path.join(save_folder, "weights.h5")
        vae.load_weights(weights_path)
        return vae


    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.reconstruction_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            reconstruction_loss = tf.reduce_mean(
                tf.reduce_sum(
                    keras.losses.binary_crossentropy(data, reconstruction), axis=(1, 2)
                )
            )
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))
            total_loss = reconstruction_loss + kl_loss
        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        self.reconstruction_loss_tracker.update_state(reconstruction_loss)
        self.kl_loss_tracker.update_state(kl_loss)
        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.reconstruction_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }
```

```python
def colorjitter(img, cj_type="b",value =-50):
    '''
    ### Different Color Jitter ###
    img: image
    cj_type: {b: brightness, s: saturation, c: constast}
    '''
    if cj_type == "b":
        # value = random.randint(-50, 50)
        value = value
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        h, s, v = cv2.split(hsv)
        if value >= 0:
            lim = 255 - value
            v[v > lim] = 255
            v[v <= lim] += value
        else:
            lim = np.absolute(value)
            v[v < lim] = 0
            v[v >= lim] -= np.absolute(value)

        final_hsv = cv2.merge((h, s, v))
        img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
        return img
    
    elif cj_type == "s":
        # value = random.randint(-50, 50)
        value = np.random.choice(np.array([-50, -40, -30, 30, 40, 50]))
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        h, s, v = cv2.split(hsv)
        if value >= 0:
            lim = 255 - value
            s[s > lim] = 255
            s[s <= lim] += value
        else:
            lim = np.absolute(value)
            s[s < lim] = 0
            s[s >= lim] -= np.absolute(value)

        final_hsv = cv2.merge((h, s, v))
        img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
        return img
    
    elif cj_type == "c":
        brightness = 10
        contrast = np.random.randint(40, 100)
        dummy = np.int16(img)
        dummy = dummy * (contrast/127+1) - contrast + brightness
        dummy = np.clip(dummy, 0, 255)
        img = np.uint8(dummy)
        return img
```

```python
def filters(img, f_type = "blur"):
    '''
    ### Filtering ###
    img: image
    f_type: {blur: blur, gaussian: gaussian, median: median}
    '''
    if f_type == "blur":
        image=img.copy()
        fsize = 3
        return cv2.blur(image,(fsize,fsize))
    
    elif f_type == "gaussian":
        image=img.copy()
        fsize = 3
        return cv2.GaussianBlur(image, (fsize, fsize), 0)
    
    elif f_type == "median":
        image=img.copy()
        fsize = 3
        return cv2.medianBlur(image, fsize)
```
